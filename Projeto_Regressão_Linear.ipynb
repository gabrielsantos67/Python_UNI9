```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from google.colab import files
```

```python
uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(filename)
```

```python
x = df[['Hours']]
y = df['Scores']
```

```python
modelo_regression = LinearRegression()
modelo_regression.fit(x, y)
```

```python
horário_especifico = 15
nota_previsto = modelo_regression.predict([[horário_especifico]])
print(f"Aluno que estuda {horário_especifico} horas, terá uma assertividade de {nota_previsto[0]:.2f}%.")
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='Hours', y='Scores', data=df)
plt.title('Relação entre Horas estudadas e assertividade das notas')
plt.show()
```

```python
import numpy as np
```

```python
hours_studied = df['Hours'].values.reshape(-1, 1)
grades = df['Scores']

model = LinearRegression()

model.fit(hours_studied, grades)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (m): {slope}")
print(f"Intercept (b): {intercept}")

hours_range = np.arange(df['Hours'].min(), df['Hours'].max(), 0.1).reshape(-1, 1)
predicted_grades = model.predict(hours_range)

plt.scatter(df['Hours'], grades, color='blue', label='Data Points')
plt.plot(hours_range, predicted_grades, color='red', label='Regression Line')
plt.title('Relação entre Horas estudadas e assertividade')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.grid(True)
plt.show()
```
